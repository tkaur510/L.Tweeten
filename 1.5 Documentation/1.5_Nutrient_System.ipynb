```python
#returns tuples (t, j) to act as a stamp for every household
marker=[]
for i in range(len(consumption["t"])):
    if (consumption["t"][i], consumption["j"][i]) not in marker:
        marker.append((consumption["t"][i], consumption["j"][i]))

#group creates tuples of households responding in acceptable units
acceptable_units={"Gram", "Kg", "Kilogram", "Litre", "Cup", "Milli Litre",  "Number", 0}
final_group=[]
consumption=consumption.fillna(0)
for n in marker:
    ind_frame1=consumption[consumption.t==n[0]]
    ind_frame2=ind_frame1[ind_frame1.j==n[1]]
    ind_measurements=ind_frame2["u"]
    ind_measurements=set(ind_measurements)
    if ind_measurements.issubset(acceptable_units):
        final_group.append(n)
        
#list of tuples every household as seen in dataset
full_IDs=[]
for j in range(len(consumption)):
    full_IDs.append((consumption["t"][j], consumption["j"][j]))

    
#to parse, we cannot use pandas values with tuoples but isin works with floats
#we now make new IDs that are divisions of original

single_value_IDs=[]
for i in range(len(full_IDs)):
    single_value_ID=full_IDs[i][1]/full_IDs[i][0]
    single_value_IDs.append(single_value_ID)

consumption["ID"]= single_value_ID

final_group_singles=[]
for i in range(len(final_group)):
    s=final_group[i][1]/final_group[i][0]
    final_group_singles.append(s)
    
marker_singles=[]
for i in range(len(marker)):
    f=marker[i][1]/marker[i][0]
    marker_singles.append(f)

#an ID is not in final group it is dropped from the new consumption table 
tf=[]
for i in range(len(single_value_IDs)):
    tf.append(({single_value_IDs[i]}).issubset(set(final_group_singles)))
consumption["tf"]=tf
consumption_new=consumption[consumption["tf"]==True]


consumption_new
```
